### Lemmatization
- Lemmatazation technique is like Stemming.
- The output we will get after Lemmatization is called ***lemma***.
- The most important difference between Stemming and Lemmatization is that Stemming does not take into account the linguistic rules or vocabulary while Lemmatization does. For this reason, Stemming is less accurate than Lemmatization.

### Lemmatization classes
- WordNetLemmatizer

### Lemmatization configuration for POS (Parts of Speech)
- noun = n (default)
- verb = v
- adjective = a
- adverb = r
- example: 



```python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet', quiet=True)
```

```python
#1) WordNetLemmatizer
corpus = ["eating","eats","eaten","writing","writes","programming","programs"]

word_net_lemmatizer = WordNetLemmatizer()
word_net_lemmatizer.lemmatize('going')

print("Lemmaztizing corpus using POS = Nouns")
for word in corpus:
    lemma = word_net_lemmatizer.lemmatize(word,pos="n")
    print(f"word = {word}, lemma = {lemma}")

print()
print("Lemmaztizing corpus using POS = Verbs")
for word in corpus:
    lemma = word_net_lemmatizer.lemmatize(word,pos="v")
    print(f"word = {word}, lemma = {lemma}")    

print()
print("Lemmaztizing corpus using POS = Adjective")
for word in corpus:
    lemma = word_net_lemmatizer.lemmatize(word,pos="a")
    print(f"word = {word}, lemma = {lemma}") 


print()
print("Lemmaztizing corpus using POS = Adverbs")
for word in corpus:
    lemma = word_net_lemmatizer.lemmatize(word,pos="r")
    print(f"word = {word}, lemma = {lemma}")    
```
