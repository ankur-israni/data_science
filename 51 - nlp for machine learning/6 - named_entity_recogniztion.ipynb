### Named Entity Recognization (NER)
- Named Entity Recognition (NER) is an NLP task that locates and classifies “entities” in text into predefined categories—commonly PERSON, ORGANIZATION, LOCATION (GPE), DATE, TIME, MONEY, PERCENT, etc.
- Example
- Text: “Satya Nadella joined Microsoft in 1992 and lives in Seattle.”
- NER output:
- Satya Nadella → PERSON
- Microsoft → ORG
- 1992 → DATE
- Seattle → GPE

### maxent_ne_chunker_tab
- maxent_ne_chunker is a Maximum Entropy–based Named Entity Recognition (NER) model used by NLTK to identify named entities such as:
    - PERSON 
    - ORGANIZATION
    - GPE (Geo-Political Entity)
    - LOCATION
    - FACILITY

### Process for Named Entity Recognition (NER)
 - 1) Tokenize the document/sentence into words
 - 2) Create the POS tags for each word in the document/sentence
 - 3) Perform the NER (Named Entity Recognition) for each word (POS tagging for these words should already be done (in previous step #2))

```python
import nltk
#nltk.download('max_ne_chunker') # Download the Chunker used for identifying = person, organization, geo-polictical-entity, location, facitlity, etc
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker_tab')

nltk.download('words')

```

```python
# Perform the Named Entity Recognition (NER)
#1) Tokenize the document/sentence into words
#2) Create the POS tags for each word in the document/sentence
#3) Perform the NER (Named Entity Recognition) for each word (POS tagging for these words should already be done (in previous step #2))

document = "The Eiffel Tower was built from 1887 to 1889  by French Engineer Gustavo"
words = nltk.word_tokenize(document)
pos_tags = nltk.pos_tag(words)
nltk.ne_chunk(pos_tags) # nltk.ne_chunk(pos_tags) performs Named Entity Recognition (NER) on a sentence that has already been part-of-speech tagged.
```
