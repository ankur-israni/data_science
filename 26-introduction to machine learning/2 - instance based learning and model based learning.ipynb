<table>
<tr>
    <th>Feature</th>
    <th>Model based training</th>
    <th>Instance based training</th>
</tr>
<tr>
    <td>Definition</td>
    <td>Learns a general model from the entire dataset and uses it for predictions</td>
    <td>Stores training examples and compares new inputs directly to them</td>
</tr>
<tr>
    <td>Analogy</td>
    <td>Learn the rules then answer</td>
    <td>Remember examples then compare</td>
</tr>    
<tr>
    <td>Examples of Algorithms using this approach</td>
    <td>
        <ul>
            <li>Linear Regression</li>    
            <li>Logistic Regression</li>    
            <li>Neural Networks</li>    
            <li>Decision Trees</li>    
            <li>SVN with learned hyperplane</li>    
        </ul>
    </td>
    <td>
        <ul>
            <li>K-Nearest Neighbours (KNN)</li>    
            <li>Kernel regression</li>    
            <li>Locally weighted regression</li>    
        </ul>
    </td>
</tr>    

<tr>
    <td>Memory usage</td>
    <td>Low - Model compresses data into parameters</td>
    <td>High - Must store most or all training instances</td>
</tr>    

<tr>
    <td>Training time</td>
    <td>High - requires model fitting</td>
    <td>Very low - almost no training phase</td>
</tr>    

<tr>
    <td>Prediction time</td>
    <td>Fast - uses a mathematical model</td>
    <td>Slow - must compare new point to many stored examples</td>
</tr>    

<tr>
    <td>Handling large datasets</td>
    <td>Effecient once trained</td>
    <td>Beccomes slow and expensive as data grows as all data instances have to be stored and sifted through to answer a query</td>
</tr>    

</table>
