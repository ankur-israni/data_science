### C) Regression vs Classification Problems
<table border="1" cellpadding="6" cellspacing="0">
  <thead>
    <tr>
      <th>Aspect</th>
      <th>Regression</th>
      <th>Classification</th>
    </tr>
  </thead>
  <tbody>
    <tr>
      <td>Output type</td>
      <td>Continuous numeric value (e.g., 3.5, 120.7)</td>
      <td>Discrete labels/classes (e.g., Yes/No, 0/1, A/B/C)</td>
    </tr>
    <tr>
      <td>Goal</td>
      <td>Predict a quantity / how much / how many</td>
      <td>Predict a category / which class</td>
    </tr>
    <tr>
      <td>Examples</td>
      <td>House price prediction, stock price, temperature</td>
      <td>Spam detection, disease diagnosis, image labeling</td>
    </tr>
    <tr>
      <td>Target variable</td>
      <td>Real-valued (e.g.,  or a numeric range)</td>
      <td>Finite set of labels (e.g., {0,1} or {cat,dog,bird})</td>
    </tr>
    <tr>
      <td>Common algorithms</td>
      <td>Linear Regression, Polynomial Regression, Random Forest Regressor, SVR</td>
      <td>Logistic Regression, Decision Trees, Random Forest Classifier, SVM, k-NN</td>
    </tr>
    <tr>
      <td>Evaluation metrics</td>
      <td>MSE, RMSE, MAE, R</td>
      <td>Accuracy, Precision, Recall, F1-score, AUC-ROC</td>
    </tr>
    <tr>
      <td>Loss functions (typical)</td>
      <td>Mean Squared Error, Mean Absolute Error</td>
      <td>Cross-Entropy Loss, Hinge Loss</td>
    </tr>
    <tr>
      <td>Decision boundary</td>
      <td>Not usually described as a boundary; models a continuous curve/line</td>
      <td>Explicit decision boundary separating classes</td>
    </tr>
    <tr>
      <td>Output interpretation</td>
      <td>Direct numeric prediction</td>
      <td>Class label (and often class probabilities)</td>
    </tr>
    <tr>
      <td>Typical use cases</td>
      <td>Forecasting sales, predicting demand, estimating costs</td>
      <td>Customer churn, fraud detection, sentiment analysis</td>
    </tr>
  </tbody>
</table>


**C1)Binary Classification**
- If there are only **two categories** for a feature (input or output feature), e.g. Pass or Fail, True or False, that is known as Binary Classification.

**C2) Multi-class Classification**
- If there are **more than two categories** for a feature (input or output), e.g. Pass, Fail, Unknown, that is known as Multi-class Classification
