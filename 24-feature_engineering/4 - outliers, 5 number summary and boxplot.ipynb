# 5 Number Summary and Box Plot
1) Minimum value
2) Maximum value
3) Mediam
4) Q1 = Quartile 1 = 25% Percentile
5) Q3 = Quartile 3 = 75% Percentile
6) IQR = Inter Quartile Range = Q3 - Q1


```python
#1) Calculate the five number summary and IQR for the given list of marks
import numpy as np
marks = [-100,45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74,150,170,180]
minimum, Q1, median, Q3, maximum = np.quantile(marks,[0,0.25,0.5,0.75,1.0])
print("minimum: ",minimum, "Q1: ",Q1, "median: ",median, "Q3: ",Q3, "maximum: ",maximum)


IQR = Q3 - Q1
print("IQR: ",IQR)


```

```python
#2) Calculate the lower and upper fences
#lower_fence - any value below this is an outlier
#upper_fence - any value above this is an outlier
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR
print("lower_fence: ", lower_fence, "upper_fence: ", upper_fence)

```

```python
#3) Creating the box plot
# The values in BLUE  color are in the normal range, 
# The circle dots are the OUTLIERS in the box plot below
import seaborn as sns
sns.boxplot(marks)
```
