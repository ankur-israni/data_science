# EDA
**Definition**
- EDA = Exploratory Data Analysis.
- EDA is the process of exploring a dataset to understand its structure, patterns, anomalies, and relationships before building any model.
- You try to “understand the data” before doing anything with it.

**Why is EDA important**
- To detect missing values
- To find outliers
- To understand data distribution
- To understand relationships between variables
- To decide which features matter
- To spot data quality issues
- To guide feature engineering and model selection

**What do we do in EDA**
1. Look at data structure
- df.head()
- df.info()
- df.describe()

2. Check missing values
- df.isna().sum()

3. Check for duplicates
- df.duplicated().sum()

4. Understand distributions
- Histograms
- Boxplots
- Density plots

5. Look at relationships
- Scatterplots
- Pair plots
- Correlation matrix (df.corr())

6. Outlier detection
- Boxplots
- Z-score
- IQR

7. Categorical analysis
- Value counts
- Bar charts

**Commonly used plots in EDA**
- Histogram
- Boxplot
- Scatter plot
- Heatmap
- Pairplot (seaborn)
- Countplot

# Co-relation
Co-relation is always between -1 to +1.
- Corelation is defined as the statistical measure that indicates the extent to which two or more variables fluctuate together.
- A positive correlation indicates that as one variable increases, the other variable tends to increase as well.
- A negative correlation indicates that as one variable increases, the other variable tends to decrease.

# Univariate and Multi-variate Analysis of data
**Univariate Analysis**
- Analysis of a Single Variable at a time
- Mean, Median, Mode
- Histogram
- Box Plot

**Multivariate Analysis**
- Analysis of multiple variables simultaneously to understand the relationships between them.
- Goal is to identify patterns, co-relations and interactions amongst variables/features
- Corelation analysis - df.corr()
- Scatter Plot
- Pair Plot
- Cross Tabulation

# Plots
**1)Box Plot**
- Purpose: Distribution + Outliers
- What it shows: Median, quartiles, IQR, outliers

**2)Scatter Plot**
- Purpose: Relationship
- What it shows: Corelation, clusters, patterns

**3)Bar Graph**
- What it shows: Count or value per category, e.g. One Bar (in bar graph) for Count = 5, Another Bar for Count = 10, etc

**4)Histogram**
- What it shows: Frequency of values in bins e.g. One Bar for values in range/bin = 1 to 10, another Bar in for values in rage 11 - 20, etc

**5)Categorical Plot**
- Purpose: Mixed purpose depending on the type of graph (count, box, scatter, etc)
- What it shows: Depends on the kind of graph being shown in Categorical plot.

**6)Pair Plot**
- Purpose: Visualiazation of multiple variables at the same time in one graph
- What it shows: Scatter plots for each pair of features + histograms

## EDA With Red Wine Data
**Overview of the Wine Dataset**
- The two datasets are related to red and white variants of the Portuguese "Vinho Verde" wine.  
- Due to privacy and logistic issues, only physicochemical (inputs) and sensory (the output) variables are available (e.g. there is no data about grape types, wine brand, wine selling price, etc.).
-These datasets can be viewed as classification or regression tasks. 
-The classes are ordered and not balanced (e.g. there are many more normal wines than excellent or poor ones). 
- Outlier detection algorithms could be used to detect the few excellent or poor wines. - Also, we are not sure if all input variables are relevant. So it could be interesting to test feature selection methods.


**Attribute Information:**
Input variables (based on physicochemical tests):
- 1 - fixed acidity
- 2 - volatile acidity
- 3 - citric acid
- 4 - residual sugar
- 5 - chlorides
- 6 - free sulfur dioxide
- 7 - total sulfur dioxide
- 8 - density
- 9 - pH
- 10 - sulphates
- 11 - alcohol

Output variable (based on sensory data):
- 12 - quality (score between 0 and 10)

**Source**
- This dataset is downloaded from www.kaggle.com
- https://www.kaggle.com/datasets/uciml/red-wine-quality-cortez-et-al-2009?resource=download

```python
# ********** Information on the dataset **********
import pandas as pd
df=pd.read_csv('./resources/winequality-red.csv')

print("printing df.head()");
print(df.head());

# df.describe()
# provides statistical summary of numerical columns in the dataframe such as count, mean, std, min, 25%, 50%, 75%, and max.
print("printing df.describe()");
print(df.describe());

# df.info()
# provides a concise summary of the dataframe, including the number of non-null entries, data types of each column, and memory usage.
print("printing df.info()");
print(df.info());

#df.shape
# returns a tuple representing the dimensionality of the dataframe (number of rows, number of columns).
print("printing df.shape");
print(df.shape);

#df.columns
# returns the column labels of the dataframe as an Index object.
print("printing df.columns");
print(df.columns);

```

```python
# ********** Unique rows in the Dataset **********
# Print unique rows in quality column (output/target variable)
print("printing unique values in quality column");
print(df['quality'].unique());
```

```python
# ********** Missing values in the dataset by column**********
# Print the missing values in the dataset by column names
print("printing missing values by column");
print(df.isnull().sum());
```

```python
# **********  Duplicate Records in the Dataset **********
# Print duplicate records in the dataset
print("total number of duplicate records in the dataset:");
print(df.duplicated().sum());

print("printing duplicate records in the dataset");
# True = duplicate record.
# False = Unique record.
print(df.duplicated());

# Return all duplicate rows in the dataframe
print("returning all duplicate rows in the dataframe");
print(df[df.duplicated()]);

# Removing the duplicate rows in the dataframe
df_without_duplicates = df.drop_duplicates()
print("printing dataframe after removing duplicate rows");
print(df_without_duplicates);

#  Shape of dataframe after removing duplicate rows
print("shape of dataframe after removing duplicate rows");
print(df_without_duplicates.shape)
```

```python
# Visualization > Correlation Matrix Heatmap
# **********  correlation between different features in the dataset **********
print("printing correlation between different features in the dataset");
print("correlation matrix:");
df.corr()

# Create a heatmap graph to visualize the correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8)) #Controls the size of the heatmap

#annot=True to show the correlation values in each cell
#cmap='coolwarm' to use a color scheme that ranges from cool colors (blues) to warm colors (reds)
#fmt=".2f" to format the correlation values to two decimal places
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
# sns.heatmap(df.corr(), annot=True)

plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Visualization > Bar Graph for Quality Counts
# **********  Bar Graph for Quality Counts **********

# Print unique rows in quality column (output/target variable)
# Quality is an OUTPUT variable with values between 0 and 10.
print("printing unique values in quality column");
print(df['quality'].unique());

# Bar graph
# This is an IMBALANCED DATASET as some Quality values have very less number of records.
print("printing bar graph for quality counts");
import matplotlib.pyplot as plt     
quality_counts = df['quality'].value_counts()
plt.bar(quality_counts.index, quality_counts.values, color='skyblue')
plt.xlabel('Quality') #Quality output column
plt.ylabel('Count')
plt.title('Bar Graph for Quality Counts')
plt.show()
```

```python
# Visualization > Bar Graphs all input features in one graph
print("printing Hist Plot  for all input features in one graph");
for column in df.columns:
   sns.histplot(df[column], kde=True)



   
```

```python
# Visualization > Hist Plot Graphs for one of the input features - alcohol
print('printing Hist Plot graph for one of the input features - alcohol');
sns.histplot(df['alcohol'], kde=True)
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.title('Hist Plot Graph for Alcohol Content')
plt.show()
```

```python
# Visualization > Pair Plot
# Used for univariate, bivariate and multivariate analysis
# It helps to visualize the distribution of single variables and the relationships between multiple variables in a dataset
print("printing pair plot for the dataset");
sns.pairplot(df)
plt.title('Pair Plot for the Dataset')
plt.show()
```

```python
# Visualization > Categorical Plot
# Used to visualize the relationship between a categorical variable and one or more numerical variables.
print("printing categorical plot for quality vs alcohol");

# kind = box means box plot
# Notice lot of OUTLIERS where quality = 5
sns.catplot(x='quality', y='alcohol', data=df, kind='box', height=6, aspect=1.5)
plt.title('Categorical Plot: Quality vs Alcohol')
plt.show()
```

```python
# Visualization > Scatter Plot
# We use numerical variables to create scatter plots to see the relationship between them.
print("printing scatter plot for alcohol vs quality");

# hue='quality' to color the points based on quality values
# - Each unique value of quality gets its own color.
# palette='viridis' to use the viridis color palette for better visualization
# - 'viridis' is a very popular color map that goes from dark purple → blue → green → yellow.
sns.scatterplot(x='alcohol', y='pH', data=df, hue='quality', palette='viridis')
plt.xlabel('Alcohol Content')
plt.ylabel('PH Level')
plt.title('Scatter Plot: Alcohol vs PH Level')
plt.show()
```
