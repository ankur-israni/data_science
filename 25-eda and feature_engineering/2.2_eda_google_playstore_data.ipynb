## Data Cleaning

```python
# 'Reviews' column has Categorical data, Get all unique values
df['Reviews'].unique()
```

```python
# Since data is already int type, we can easily convert from Object type to int type. No need to Encode (OHE etc)
# This will however give an error because ONE value or 'Review' is not numeric (i.e. 3.0M)
# Shown in steps below
df['Reviews'].astype(int)
```

```python
df['Reviews'].str.isnumeric().sum()
```

```python
# Find non numeric rows in the column 'Reviews' 
# Only one row, Review = 3.0M
df[~df['Reviews'].str.isnumeric()]
```

```python
# Make a copy of the origional dataframe
df_copy=df.copy()
```

```python
# From the df_copy dataframe, delete the index number 10472. 
# The index number is like the serial number, not an actual column name
df_copy=df_copy.drop(df_copy.index[10472])
```

```python
# Find all the non-numeric values in the column 'Review' in the dataframe df_copy
# After having deleted the only index (index = 10472 with Review = 3M, this should give zero rows)
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
## Convert Review Datatype to int
df_copy['Reviews']=df_copy['Reviews'].astype(int)
```

```python
# This should show 'Reviews' as type (int)
df_copy.info()
```

```python
# 'Size' column also has Classification data.
# Print all the unique values of this column
df_copy['Size'].unique()
```

```python
# Find the count of rows where Size = null - Zero rows in this dataset
df_copy['Size'].isnull().sum()
```

```python
# Convert values of 'Size' into numeric values
# Replace M (million) with '000' in the column = Size
df_copy['Size']=df_copy['Size'].str.replace('M','000')

# Replace K (kilo) with '' (blank string) in the column = Size
df_copy['Size']=df_copy['Size'].str.replace('k','')

# Replace 'Varies with device' with 'np.nan' in the column = Size
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)

# Convert 'Size' column to float datatype
df_copy['Size']=df_copy['Size'].astype(float)
```

```python
# Print the updated values in 'Size' column
df_copy['Size']
```

```python
# Installs column also has Classification type values
# Print all the unique values in this column
df_copy['Installs'].unique()
```

```python
# Price column also has Classification type values
# Print all the unique values in this column
df_copy['Price'].unique()
```

```python
# Replace characters like + , $ from the column 'Installs' and 'Price' with '' (blank string)
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(item,'')
```

```python
# Print the value of 'Price' column, it should be without any special characters (+ , $)
df_copy['Price'].unique()
```

```python
# Print the value of 'Installs' column, it should be without any special characters (+ , $)
df_copy['Installs'].unique()
```

```python
# Convert the 'Installs' and 'Price' coluns into 'int' and 'float' datatypes respectively
df_copy['Installs']=df_copy['Installs'].astype('int') # Int since it does not have any decimal values
df_copy['Price']=df_copy['Price'].astype('float') # Float since this has decimal values
```

```python
df_copy.info()
```

```python
## Handlling Last update feature
# 'Last updated' also has Classification type records.
df_copy['Last Updated'].unique()
```

```python
# Convert the values of 'Last Updated' to 'datetime' datatype
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])

# Create new columns 'Day','Month','Year' from 'Last Updated'
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year
```

```python
# Notice the datatype of 'Last Updated' is datetime
# And the datatype of 'Day', 'Month' and 'Year' is (int)
df_copy.info()
```

```python
# Notice the values of 'Day','Month' and 'Year' are (int) values
df_copy.head()
```

## Explore Data

```python
# Find all numeric features (i.e. columns)
# df_copy.columns = all columns
# df_copy[feature].dtype != 'O']
# - dtype = datatype
# dtype!= O (Oh not zero), O = Not Object type( i.e. Not String, Categorical data, etc)    
numeric_features = [feature for feature in df_copy.columns if df_copy[feature].dtype != 'O']

# dtype == O (Categorical data) (i.e. Is One of String, Categorical data, etc)
categorical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

## 3.2 Feature Information
1. App :- Name of the App
2. Category :- Category under which the App falls.
3. Rating :- Application's rating on playstore
4. Reviews :- Number of reviews of the App.
5. Size :- Size of the App.
6. Install :- Number of Installs of the App
7. Type :- If the App is free/paid
8. Price :- Price of the app (0 if it is Free)
9. Content Rating :- Appropiate Target Audience of the App.
10. Genres:- Genre under which the App falls.
11. Last Updated :- Date when the App was last updated
12. Current Ver :- Current Version of the Application
13. Android Ver :- Minimum Android Version required to run the App

```python
# For all the Categorial Columns (e.g. App, Category, Installs, etc. find out what percentage of data each category has)
# - e.g. In the Column 'App' Robox is 0.08 % of all the different values in 'App' column
# normalize=True changes what the value_counts return
# - when normilize=false or missing, then output = 5,3,4 etc
# - when normilize=true , then output = relative proportions instead of counts - they add up to 1.0 e.g. - 0.5, 0.3. 0.2. if you multiply by 100, you will get the percentage.

for col in categorical_features:
    print(df[col].value_counts(normalize=True)*100)
    print('---------------------------')
```

```python
# Univariate Analysis for All Columns having Numerical data
# kdeplot = Kernel Density(Y Axis)
# Numerical Feature/Column (X-Axis)

# Observation
# - A few small data points pull the tail to the Left
# - The following graphs below are Left Skewed
#  - Rating
#  - Last Updated
#  - Year

# - The following graphs below are Right Skewed
#   - Reviews
#   - Size
#   - Installs
#   - Price


plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df_copy[numeric_features[i]],shade=True, color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

```python
# Plot of Categorical Columns
# Whenever we have Categorical data, we can use Count Plot.
plt.figure(figsize=(20, 15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)


category = [ 'Type', 'Content Rating']

# This loop will run 2 times, 
# 1st time = Will create a plot for 'Type' and its count
# 2nd time = Will create a plot for 'Content Rating' and its count
# palette=Set 2 (will choose from a predefined set of colors for Plot)
# plt.subplot(2, 2, i+1), 2X2 graph = Total 4 spaces, we are only using two spaces for 'Type' and ' Content Rating'
# First pass of for Loop : i+1 = 1 (when i=0) - This is the position for 1st plot = Type
# Second pass of for Loop : i+1 =2 (when i=1) - This is the position for 2nd plot = Content Rating
# plt.xtics(rotation=45) = Rotates X-value labels by 45 degrees for better readability.

for i in range(0, len(category)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=df[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout() 

    
```

## Which Category has largest number of installations??

```python
df_cat_installs = df_copy.groupby(['Category'])['Installs'].sum().sort_values(ascending = False).reset_index()
# print(df_cat_installs)

df_cat_installs.Installs = df_cat_installs.Installs #1000000000# converting into billions
df2 = df_cat_installs.head(10)
plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x = 'Installs' , y = 'Category' , data = df2 )
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```

## Insights
1. Out of all the categories "GAME" has the most number of Installations.
2. With almost 35 Billion Installations GAME is the most popular Category in Google App store

## What are the Top 5 most installed Apps in Each popular Categories ??

```python
dfa = df_copy.groupby(['Category' ,'App'])['Installs'].sum().reset_index()
dfa = dfa.sort_values('Installs', ascending = False)
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = dfa[dfa.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,x= 'Installs' ,y='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)
    
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```

## Insights
- Most popular game is Subway Surfers.
- Most popular communication app is Hangouts.
- Most popular productivity app is Google Drive.
- Most popular social app is Instagram.

## How many apps are there on Google Play Store which get 5 ratings??

```python
rating = df_copy.groupby(['Category','Installs', 'App'])['Rating'].sum().sort_values(ascending = False).reset_index()

toprating_apps = rating[rating.Rating == 5.0]
print("Number of 5 rated apps",toprating_apps.shape[0])
toprating_apps.head(1)
```
