### 2) Architecture of a Perceptron
- <image src="./resources/images/neural_networks_1.png" width=600 height=500/>
- (src = ./resources/images/perceptron_1.drawio)
- ***Weights***
    - Each connection between the input(s) neurons has an associated weight (w1,w2, w3...)
    - A Positive Weight increases the influence of an input.
    - A Negative Weight decreases or inversely affects the input.
    - Weights are initially ***initialized randomly***.
- ***Bias***
    - If the weights when initialized are randomly initialized to 0, W1X1 = 0 (in Step 1), where W1 = weight 1, X1 = Input 1
    - To avoid from this happening we add a Bias with some initial value so that the whole neuron's value is not 0 (0 = deactivated)
    - The Bias is typically Noise. 
    - Noise basically means that data is totally different from the rest of the data - But our model should be capable enough to handle that as well.
- ***Perceptron is a Linear classifier***
    -  Step 1's equation (below) is exactly similar to the Equation of a Line.
    - This means that Perceptron is a Linear Classifier
    - <image src="./resources/images/perceptron_linear_classifier_1.png" width=400 height=300/>

### 4) Perceptron Models
- Perceptron Models are of two types
  - 1) Single Layered Perceptron Models = Perceptron
  - 2) Multi Layered Perceptron Models = ANN (Artificial Neural Networks)
- <image src="./resources/images/perceptron_models.png" width=700 height=300>
- Single Layered Perceptron 
  - Are for Simpler Classification Problems
  - Problems which can be solved Linearly
  
- Multi Layered Perceptron = ANN (Artificial Neural Network)  
  - Are for solving more Complex problems
  - Problems that cannot be solved Linearly

- <image src="./resources/images/perceptron_vs_ann_1.png" width=600 height=500/>  

  - Standard techniques used in Multi-Layered Perceptron Model
    - 1) Forward propogration
    - 2) Backward propogation
    - 3) Loss functions
    - 4) Activation functions
    - 5) Optimizers
