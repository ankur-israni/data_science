### Deep Learning
- Overview
    - Machine Learning is a subset of Artificial Intelligence.
    - Deep Learning is a subset of Machine Learning.
    - Deep Learning can be:-
        - 1) Supervised Deep Learning
        - 2) UnSupervised Deep Learning

### Why is Deep Learning popular
- 1) Large data availaible: Large amount of data generated by social networking and other sites(netflix, facebook) over the years.
    - This data needs analysis by companies to increase customer behaviours so revenue can be increased.
- 2) Improvement in Hardware:
    - Hardware needed to run these ML and DL algorithms has improved considerably - by NVidia
- 3) Usecases avaialible across different industries
    - ML and DL usecases are availaible across many different domains - medical, ecommerce, retail, marketing, etc  
- 4) Popular frameworks by big companies availaible for DL
    - Frameworks like Pytorch (by Facebook) and TensorFlow (by Google) makes it easy to perform Deep Learning       

### Components of Deep Learning
- 1) ANN
    - ANN = Artificial Neural Networks
    - ANN is used to solve
        - a) Classification problems
        - b) Regression problems
- 2) CNN
    - CNN = Convolutional Neural Network
    - Helps us in solving Classification problems
    - Input data is mainly:-
        - a) Images
        - b) Video frames
    - Variants of CNN are 
        - a) RCNN
        - b) Masked RCNN
        - c) DetectRon
        - d) Yolo v5, v6, v7
    - Computer Vision has 3 main tasks. These sub-tasks uses CNN techniques.
        - a) Image Classification
        - b) Object Detection
        - c) Image Segmentation
- 3) RNN     
    - RNN = Recurrent Neural Network
    - RNN is used when the input data is:-
        - a) Text data
        - b) Time Series data
            - Time Series data 
                - e.g. Forecasting sales data
                - Sequence of data prediction wherein tomorrows data is dependent on yesterday's sequences
        - c) Variants of RNN
            - LSTM RNN
            - GRU RNN
            - Bidirectional LSTM RNN
            - Encoder Decoder
            - Transformer
            - BERT
